```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
n = int(input())
phone_book = {}
# input
for i in range(n):
    name, phoneNumber = input().split()
    phone_book[name] = phoneNumber

while True:
    try:
        query = input()
    except:
        break

    result = query + '=' + phone_book[query] if query in phone_book else 'Not found'
    print(result)


# if find or erro
```
